## VISUALIZATION

```python
df['quality'].value_counts().plot(kind = 'bar', color = 'blue')
plt.xlabel('Wine Quality')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df['alcohol'], kde =True)
plt.show()
```

```python
sns.catplot(x = 'quality', y = 'alcohol', data=df, kind ='box')
plt.show()
```

```python
sns.catplot(x = 'quality', y = 'volatile acidity', data=df, kind = 'box')
plt.show()
```

```python
sns.scatterplot(x = 'alcohol', y = 'pH', data=df)
plt.show()
```

```python
sns.scatterplot(x = 'alcohol', y = 'pH', data=df, hue = 'quality')
plt.show()
```
